```python
import numpy as np
import matplotlib.pyplot as plt
t1=[10*np.sin(10*np.pi*t) for t in np.linspace(0,3,1000)]
t2=[12*np.sin(1*np.pi*t) for t in np.linspace(0,3,1000)]
t3=[21*np.sin(20*np.pi*t) for t in np.linspace(0,3,1000)]
SVM = [i+j+k for i,j,k in zip(t1,t2,t3)]
m=100
l=m-1/m
xnn=list(np.zeros(m-1))+list(SVM)
yn = [l*(sum(xnn[k-m+1:k]))+(1-l)*xnn[k] for k in np.arange(m-1,len(xnn))]

S=np.fft.fft(SVM)
plt.plot(SVM)
print(SVM)
plt.plot(abs(S[0]))
plt.plot(yn)
plt.show()
```
